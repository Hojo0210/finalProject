```python
import pandas as pd
import plotly.express as px
import hvplot.pandas
from sklearn.preprocessing import StandardScaler, MinMaxScaler
import plotly.figure_factory as ff
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
from pathlib import Path
```

```python
file_path = "D:/finalProject/MLtest/Master_data_ML.csv"
M_df = pd.read_csv(file_path)
M_df.head()
```

```python
bins = [40000, 50000, 60000, 70000, 80000, 90000, 100000, 110000, 120000, 130000, 140000, 150000,
       160000, 170000, 180000, 190000, 200000, 210000, 220000, 230000,
       240000, 250000, 260000, 270000, 280000, 290000, 300000, 310000,
       320000, 330000, 340000, 350000, 360000, 370000, 380000, 390000, 
       400000, 410000, 420000, 430000, 440000, 450000, 460000, 470000,
       480000, 490000, 500000, 510000, 520000, 530000, 540000, 550000,
       560000, 570000, 580000, 590000, 600000, 610000, 620000, 630000, 
       640000]
```

```python
M_df['SFMHVrange'] = pd.cut(M_df["singlefamilymedianhomevalue"], bins)
```

```python
M_df.SFMHVrange = pd.Categorical(M_df.SFMHVrange)
M_df["SFMHVclass"] = M_df.SFMHVrange.cat.codes
M_df
```

```python
pd.set_option('display.max_rows', None, 'display.max_columns', None)
```

```python
columns = ['county', 'year', 'totalpopulation', 'average_yearly_wage',
       'lowtiermedianhomevalue', 'hightiermedianhomevalue',
       'singlefamilymedianhomevalue', 'value_inches', 'murder', 'rape',
       'robbery', 'aggravated_assault', 'burglary', 'larceny',
       'motor_vehicle_theft', 'index_per_100k', 'employment', 'unemployment',
       'labor_force', 'unemployment_rate', 'county_code', 'SFMHVclass' ]
target = ['SFMHVrange']
```

```python
X = pd.get_dummies(M_df.drop(columns = 'SFMHVrange'))
y = M_df['SFMHVrange']
y = y.astype('str')
X.dtypes
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=50)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
X_train2, X_test2, y_train2, y_test2 = train_test_split(X, y, random_state=50, train_size=0.80)
```

```python
print(X_train2.shape)
print(X_test2.shape)
print(y_train2.shape)
print(y_test2.shape)
```

```python
scaler = StandardScaler()
X_scaler = scaler.fit(X_train)
```

```python
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
X_train_scaled
```

```python
rf_model = RandomForestClassifier(n_estimators=100, random_state=50) 
```

```python
rf_model = rf_model.fit(X_train_scaled, y_train)
```

```python
predictions = rf_model.predict(X_test_scaled)
predictions
```

```python
cm = confusion_matrix(y_test, predictions)

cm_df = pd.DataFrame(
    cm)

cm_df
```

```python
acc_score = accuracy_score(y_test, predictions)
```

```python
print("Confusion Matrix")
display(cm_df)
print(f"Accuracy Score : {acc_score}")
print("Classification Report")
print(classification_report(y_test, predictions))
```
